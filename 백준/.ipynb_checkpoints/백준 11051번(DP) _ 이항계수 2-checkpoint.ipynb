```python
a,b = map(int, input().split())
b = min(b, a-b)
n=1
for i in range(b):
    n*= a-i
for i in range(1, b+1):
    n//=i
print(n%10007)
```

```python
a,b = map(int, input().split())
b = min(b, a-b)
dp = [[0] + [0 for _ in range(k)] for k in range(a+1)]
for i in range(1,a+1):
    for j in range(len(dp[i])):
        if(j == 0) or i==j:
            dp[i][j] = 1
        else:
            dp[i][j] = (dp[i-1][j-1] + dp[i-1][j])%10007
print(dp[a][b])
```
